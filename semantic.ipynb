```python
import spacy
nlp = spacy.load("en_core_web_md")
```

```python
word1 = nlp("cat")
word2 = nlp("monkey")
word3 = nlp("banana")

print("cat vs monkey:", word1.similarity(word2))
print("banana vs monkey:", word3.similarity(word2))
print("banana vs cat:", word3.similarity(word1))
```

Cat and monkey are the most similar words which makes sense because both are animals. Banana and monkey are also closely associated becuase monkeys eat bananas. Banana and cat that the lowest similarity.

```python
# My own example
w1 = nlp("screwdriver")
w2 = nlp("hammer")
w3 = nlp("bird")

print("screwdriver vs hammer:", w1.similarity(w2))
print("screwdriver vs bird:", w1.similarity(w3))
print("hammer vs bird:", w2.similarity(w3))
```

Screwdrivers and hammers are both tools but they are not the same tool so should be closely associated but not identical. Both having little relation to a bird makes sense.

```python
# Compare en_core_web_sm to en_core_web_md
nlp_sm = spacy.load("en_core_web_sm")

cat_md = nlp("cat")
monkey_md = nlp("monkey")
banana_md = nlp("banana")

cat_sm = nlp_sm("cat")
monkey_sm = nlp_sm("monkey")
banana_sm = nlp_sm("banana")

print("\nWord similarity (en_core_web_md):")
print("cat vs monkey:", cat_md.similarity(monkey_md))
print("banana vs monkey:", banana_md.similarity(monkey_md))

print("\nWord similarity (en_core_web_sm):")
print("cat vs monkey:", cat_sm.similarity(monkey_sm))
print("banana vs monkey:", banana_sm.similarity(monkey_sm))
```

When using the en_core_web_sm model, spaCy produced a warning indicating that the model does not include pre-trained word vectors. It recommended using a larger model.
