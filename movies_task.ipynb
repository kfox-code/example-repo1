```python
import spacy
nlp = spacy.load("en_core_web_md")
nlp_sm = spacy.load("en_core_web_sm")
```

```python
def load_movies(filepath):
    movies = []
    with open(filepath, "r", encoding="utf-8") as f:
        for line in f:
            title, desc = line.split(":", 1)
            movies.append((title.strip(), desc.strip()))
    return movies

movies = load_movies("movies.txt")
```

```python
planet_hulk_desc = (
    "Will he save their world or destroy it? When the Hulk becomes too dangerous "
    "for the Earth, the Illuminati trick Hulk into a shuttle and launch him into space "
    "to a planet where the Hulk can live in peace. Unfortunately, Hulk lands on the "
    "planet Sakaar where he is sold into slavery and trained as a gladiator."
)

def recommend_movie(description, movies):
    target_doc = nlp(description)
    best_title = None
    best_score = -1

    for title, desc in movies:
        if title.lower() == "planet hulk":
            continue
        score = nlp(desc).similarity(target_doc)
        if score > best_score:
            best_score = score
            best_title = title

    return best_title, best_score

recommend_movie(planet_hulk_desc, movies)
```
